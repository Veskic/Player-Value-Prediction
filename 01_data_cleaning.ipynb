```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/players_21.csv')
print("Original shape:", df.shape)
df.head(3)
```

```python
pd.set_option('display.max_columns', None)
print("Columns in dataset:", df.columns.tolist())
```

```python
## Select features and target
features = [
    'age', 'height_cm', 'weight_kg',
    'pace', 'shooting', 'passing',
    'dribbling', 'defending', 'physic'
]

columns = ['short_name'] + features + ['value_eur']
df = df[columns]
```

```python
df_clean = df.dropna()
df_clean = df_clean[df_clean['value_eur'] > 0]
```

```python
df_clean.head()
df_clean.describe()
```

```python
df_clean.to_csv('Data/cleaned_players_21.csv', index=False)
```
